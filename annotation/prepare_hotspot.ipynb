```python
import numpy as np
import pandas as pd
import os
```

```python
df_hier = pd.read_csv('../data/hier.csv')
df_hier.head()
```

```python
num_nodes = [0,0,0,0,0,0]

for i in range(0, 6):
    length = df_hier[f'level_{i}'].unique()
    num_nodes[i] = length.size
    print(num_nodes[i])
```

```python
df_depmap_cmap = pd.read_csv('../data/depmap_cmap_hotspots.csv')
df_depmap_cmap.head()
```

```python
nodes = df_depmap_cmap[['complex', 'source']].drop_duplicates()
nodes.head()
```

```python
def getNodes(start, end):
    return nodes[(nodes['complex'] >= start) & (nodes['complex'] <= end)]

def normalize(node, level):
    node['complex'] = node['complex'] - sum(num_nodes[level+1:]) + sum(num_nodes[:level])

l5 = getNodes(0, 656)
l4 = getNodes(657, 808)
l3 = getNodes(809, 861)
l2 = getNodes(862, 875)
l1 = getNodes(876, 877)
l0 = getNodes(878, 878)

normalize(l5, 5)
normalize(l4, 4)
normalize(l3, 3)
normalize(l2, 2)
normalize(l1, 1)
normalize(l0, 0)
```

```python
export = pd.concat([l0, l1, l2, l3, l4, l5])
export.head()
```
